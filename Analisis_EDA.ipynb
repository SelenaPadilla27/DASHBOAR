```python
df_2010_2017 = pd.read_csv("df_cordoba_2010_2017.csv", sep=",", dtype=str)
df_2018_2021 = pd.read_csv("df_cordoba_2018_2021.csv", sep=",", dtype=str)
```

```python
f"cantidad de columnas en df1 {len(df_2010_2017.columns)} cantidad de columnas en df2 {len(df_2018_2021.columns)}"
```

```python
for i in df_2010_2017.columns:
    for j in df_2018_2021.columns:
        if(i == j):
            print(f"columnas de df1 {i}  se encuentra en df2 {j}")
            break
```

```python
#Conociendo que todas las columnas coinciden tanto en nombre como en numero total de columnas, podemos concatenar
consolidado_2010_2021 = pd.concat([df_2010_2017,df_2018_2021])
```

```python
f"La sumatoria de los df de entrada es: {df_2010_2017.shape[0]+df_2018_2021.shape[0]} El total luego de concatenar es: {consolidado_2010_2021.shape[0]}"
```
